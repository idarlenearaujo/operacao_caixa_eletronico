```python
# version 1.0
# Main chamada principal
if __name__ == '__main__':
    print('Início do sistema Bancário!\n')

historico_conta = [] # Armazenando historico de saques/depositos
valor_conta = 0 # inicializando valor da conta
opcao = -1 
saques_realizados = 0

def depositar_cash(valor_conta): # função para depositar na conta, existe verificação de valores negativos
    
    try:
        valor_deposito = float(input("\nDigite o valor que deseja depositar: ")) # recebe o valor

        if valor_deposito > 0: # verifica e se maior que 0 retorna valor para conta
            return valor_deposito
    
    except:
        print('\nENTRADA INVÁLIDA!!!\nValor esperado: > 0.00\nInforme um valor válido.') # se valor 0 ou inferior, retorna uma mensagem e o valor 0 para conta

def sacar_cash(valor_conta, saques_realizados): # função para sacar na conta, existe a verificação se o valor a ser sacado é possível
    
    if saques_realizados < 3:
        
        try:
            valor_saque = float(input("\nDigite o valor que deseja sacar: ")) # recebe valor

            if 0 < valor_saque <= 500: # verifica se o saque é maior que 0 e se número de saques é inferior a 3 e se o valor é menor ou igual a 500,00
                if valor_conta >= valor_saque: # verifica se existe saldo
                    return valor_saque 
                
                print("\nSALDO INSUFICIENTE!!!") # se conta não tem saldo suficiente para o saque, retorna mensagem de erro
            
        except:
            print('\nO valor informado não deve ser zero, valor negativo ou superior a R$ 500.00.\nInforme um valor válido.')
    
    else:
        print('\nVocê atingiu o limite de 3 saques.')                                                                                                                                                    
        
    return 0

def extrato_cash(valor_conta, historico_conta): # mostra extrato de todas as movimentação da conta ativa

    for dados in historico_conta:
        print(dados)

    print(f"\nValor em conta atualizado: R$ {valor_conta:.2f}") # Mostra valor em conta

# Deposito
# Saque
# Extrato
def menu():
    print("\nInforme a opção desejada:\nDigite 1: Depósito\nDigite 2: Saque\nDigite 3: Extrato\nDigite 0: Sair\n")

while opcao != 0: # Loop simulando um caixa eletrônico

    menu()

    try:
        opcao = int(input())

        if opcao == 1:
            
            valor_deposito = depositar_cash(valor_conta)
            valor_conta += valor_deposito
            historico_conta.append(f"Valor de + R$ {valor_deposito:.2f} foi creditado")
            print(f'Valor R$ {valor_deposito:.2f} foi depositado.')

        elif opcao == 2:
            
            valor_sacado = sacar_cash(valor_conta, saques_realizados)
            
            if valor_sacado > 0:
                valor_conta -= valor_sacado
                historico_conta.append(f"Valor de - R$ {valor_sacado:.2f} foi debitado")
                print(f'Valor R$ {valor_sacado:.2f} foi sacado.')
                saques_realizados += 1
                
        elif opcao == 3:
            print('\n-------------------------------------')
            print('--------------- EXTRATO -------------')
            extrato_cash(valor_conta, historico_conta)
            print('\n-------------------------------------')
            
        elif opcao == 0:
            print('\nSaindo da sua conta!\n\n# # # # # # # # # # # # # # # # # # #')

        else: 
            print('\nValor digitado inválido!')

    except:
        print('\nValor digitado inválido!')
```
